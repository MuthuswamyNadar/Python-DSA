# ---------------------- Theory ------------------------------

 # 1) What is Boosting in Machine Learning?



Boosting is an ensemble technique in machine learning that combines multiple weak learners sequentially to create a strong model by iteratively focusing on correcting the errors made by previous models.

# 2. How does Boosting differ from Bagging? 

   Boosting differs from Bagging because Boosting builds models sequentially, where each model focuses on improving errors made by its predecessor, while Bagging trains models independently in parallel and combines their outputs.


# 3. What is the key idea behind AdaBoost? 
 The key idea behind AdaBoost (Adaptive Boosting) is to assign higher weights to misclassified instances so that subsequent models focus more on correcting these difficult cases.


# 4. Explain the working of AdaBoost with an example.  
   AdaBoost works by training a weak model on the dataset, calculating its errors, and increasing the weights of misclassified samples. This process is repeated iteratively, and the final prediction is based on a weighted majority vote of all weak models. For instance, if classifying spam emails, AdaBoost adjusts focus on incorrectly classified emails until accuracy improves.



# 5. What is Gradient Boosting, and how is it different from AdaBoost?  
   Gradient Boosting is a sequential boosting method that minimizes a loss function using gradient descent to adjust predictions. Unlike AdaBoost, Gradient Boosting allows more flexibility and can handle complex tasks effectively.



# 6. What is the loss function in Gradient Boosting?  
   The loss function in Gradient Boosting measures the discrepancy between predicted and actual values. Examples include mean squared error for regression tasks and log loss for classification.

# 7. How does XGBoost improve over traditional Gradient Boosting?  
   XGBoost improves traditional Gradient Boosting by introducing regularization techniques, distributed computing for faster training, and efficient handling of sparse data, resulting in enhanced accuracy and scalability.



# 8. How does regularization help in XGBoost?  
   Regularization in XGBoost prevents overfitting by applying techniques like L1 (Lasso) and L2 (Ridge) regularization, ensuring the model generalizes well to unseen data.



# 9. What are some hyperparameters to tune in Gradient Boosting models?  
   Key hyperparameters include learning rate, number of trees, maximum depth, subsample ratio, and minimum child weight. Tuning these can optimize model performance.



# 10. What is the difference between XGBoost and CatBoost?  
   XGBoost focuses on speed and flexibility, while CatBoost is efficient for datasets with categorical features and requires less preprocessing.



# 11. Why is CatBoost efficient for categorical data?  
   CatBoost uses ordered boosting and handles categorical features natively, eliminating the need for extensive encoding.



# 12. What are some real-world applications of Boosting techniques?  
   Boosting techniques are widely applied in fraud detection, risk assessment, customer churn prediction, healthcare diagnostics, recommendation systems, and NLP tasks like sentiment analysis.



# 13. What is the concept of Feature Importance in Boosting?  
   Feature Importance measures how much each feature contributes to the model's predictive power, aiding in feature selection and understanding key drivers of predictions.

# 1. Train an AdaBoost Classifier on a sample dataset and print model accuracy


```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load Iris dataset
data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train AdaBoost Classifier
model = AdaBoostClassifier(n_estimators=50, random_state=42)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Model Accuracy: {accuracy:.2f}")
```

# 4. Train an AdaBoost Classifier with different numbers of estimators and compare accuracy

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load Iris dataset
data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Compare accuracies for different estimators
estimators = [10, 50, 100, 200]
accuracies = []
for n in estimators:
    model = AdaBoostClassifier(n_estimators=n, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

# Plot comparison
plt.plot(estimators, accuracies, marker='o')
plt.title("Effect of Estimators on AdaBoost Accuracy")
plt.xlabel("Number of Estimators")
plt.ylabel("Accuracy")
plt.grid()
plt.show()
```

# 5. Train an AdaBoost Classifier and analyze the effect of different learning rates

```python
from sklearn.ensemble import AdaBoostClassifier
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
import matplotlib.pyplot as plt

# Load Iris dataset
data = load_iris()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Compare accuracies for different learning rates
learning_rates = [0.1, 0.5, 1.0, 1.5]
accuracies = []
for lr in learning_rates:
    model = AdaBoostClassifier(learning_rate=lr, random_state=42)
    model.fit(X_train, y_train)
    y_pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

# Plot comparison
plt.plot(learning_rates, accuracies, marker='o', color='orange')
plt.title("Effect of Learning Rates on AdaBoost Accuracy")
plt.xlabel("Learning Rate")
plt.ylabel("Accuracy")
plt.grid()
plt.show()
```

# 8. Train a Gradient Boosting Regressor and plot learning curves

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split

# Generate regression data
X, y = make_regression(n_samples=500, n_features=10, noise=0.1, random_state=42)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Gradient Boosting Regressor
model = GradientBoostingRegressor(n_estimators=100, random_state=42)
training_errors = []
validation_errors = []

for i in range(1, len(X_train) + 1):
    model.fit(X_train[:i], y_train[:i])
    training_errors.append(model.score(X_train[:i], y_train[:i]))
    validation_errors.append(model.score(X_test, y_test))

# Plot learning curves
plt.plot(training_errors, label="Training Score")
plt.plot(validation_errors, label="Validation Score")
plt.title("Learning Curves - Gradient Boosting")
plt.xlabel("Training Examples")
plt.ylabel("Score")
plt.legend()
plt.grid()
plt.show()
```

# 9. Train a Gradient Boosting Classifier and visualize the ROC curve

```python
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.metrics import roc_curve, auc
import matplotlib.pyplot as plt

# Load Breast Cancer dataset
data = load_breast_cancer()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Gradient Boosting Classifier
model = GradientBoostingClassifier(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# Compute ROC curve
y_prob = model.predict_proba(X_test)[:, 1]
fpr, tpr, _ = roc_curve(y_test, y_prob)
roc_auc = auc(fpr, tpr)

# Plot ROC curve
plt.plot(fpr, tpr, label=f"ROC Curve (AUC = {roc_auc:.2f})")
plt.title("ROC Curve - Gradient Boosting Classifier")
plt.xlabel("False Positive Rate")
plt.ylabel("True Positive Rate")
plt.legend()
plt.grid()
plt.show()
```

# 10. Train an XGBoost Classifier on a dataset and compare accuracy with Gradient Boosting

```python
from xgboost import XGBClassifier
from sklearn.ensemble import GradientBoostingClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score

# Load sample dataset
data = load_digits()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train XGBoost Classifier
xgb_model = XGBClassifier(n_estimators=100, random_state=42, use_label_encoder=False, eval_metric='mlogloss')
xgb_model.fit(X_train, y_train)
xgb_pred = xgb_model.predict(X_test)
xgb_accuracy = accuracy_score(y_test, xgb_pred)

# Train Gradient Boosting Classifier
gb_model = GradientBoostingClassifier(n_estimators=100, random_state=42)
gb_model.fit(X_train, y_train)
gb_pred = gb_model.predict(X_test)
gb_accuracy = accuracy_score(y_test, gb_pred)

print(f"XGBoost Accuracy: {xgb_accuracy:.2f}")
print(f"Gradient Boosting Accuracy: {gb_accuracy:.2f}")
```

# 13. Train an XGBoost Regressor and tune the learning rate using GridSearchCV

```python
from xgboost import XGBRegressor
from sklearn.model_selection import GridSearchCV
from sklearn.datasets import make_regression
from sklearn.metrics import mean_squared_error

# Generate regression data
X, y = make_regression(n_samples=500, n_features=10, noise=0.1, random_state=42)

# Define GridSearchCV parameters
param_grid = {'learning_rate': [0.01, 0.1, 0.2, 0.3]}
xgb_model = XGBRegressor(n_estimators=100, random_state=42)

# Perform GridSearchCV
grid_search = GridSearchCV(estimator=xgb_model, param_grid=param_grid, scoring='neg_mean_squared_error', cv=3)
grid_search.fit(X, y)
best_model = grid_search.best_estimator_

# Evaluate using MSE
y_pred = best_model.predict(X)
mse = mean_squared_error(y, y_pred)
print(f"Best Learning Rate: {grid_search.best_params_['learning_rate']}")
print(f"Mean Squared Error: {mse:.2f}")
```

# 15. Train a CatBoost Classifier and evaluate using F1-Score

```python
from catboost import CatBoostClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# Load sample dataset
data = load_digits()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train CatBoost Classifier
model = CatBoostClassifier(iterations=100, verbose=False)
model.fit(X_train, y_train)

# Predict and evaluate
y_pred = model.predict(X_test)
f1 = f1_score(y_test, y_pred, average='weighted')
print(f"F1-Score: {f1:.2f}")
```

# 16. Train a CatBoost Classifier and plot the confusion matrix

```python
from catboost import CatBoostClassifier
from sklearn.datasets import load_digits
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix
import seaborn as sns
import matplotlib.pyplot as plt

# Load sample dataset
data = load_digits()
X, y = data.data, data.target
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train CatBoost Classifier
model = CatBoostClassifier(iterations=100, verbose=False)
model.fit(X_train, y_train)

# Predict and generate confusion matrix
y_pred = model.predict(X_test)
cm = confusion_matrix(y_test, y_pred)

# Plot confusion matrix
sns.heatmap(cm, annot=True, fmt="d", cmap="Blues")
plt.title("Confusion Matrix - CatBoost")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()
```

# 17. Train a CatBoost Classifier on an imbalanced dataset and compare performance with class weighting

```python
from catboost import CatBoostClassifier
from sklearn.datasets import make_classification
from sklearn.model_selection import train_test_split
from sklearn.metrics import f1_score

# Generate imbalanced dataset
X, y = make_classification(n_samples=1000, weights=[0.9, 0.1], n_classes=2, random_state=42)

# Split dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train CatBoost without class weighting
model_default = CatBoostClassifier(iterations=100, verbose=False)
model_default.fit(X_train, y_train)
y_pred_default = model_default.predict(X_test)

# Train CatBoost with class weighting
model_weighted = CatBoostClassifier(iterations=100, class_weights=[0.1, 0.9], verbose=False)
model_weighted.fit(X_train, y_train)
y_pred_weighted = model_weighted.predict(X_test)

# Compare F1 scores
f1_default = f1_score(y_test, y_pred_default)
f1_weighted = f1_score(y_test, y_pred_weighted)

print(f"F1-Score (Default Model): {f1_default:.2f}")
print(f"F1-Score (Weighted Model): {f1_weighted:.2f}")
```
