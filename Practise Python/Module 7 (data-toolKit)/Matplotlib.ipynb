```python
import matplotlib.pyplot as plt  # Importing matplotlib.pyplot as plt

# Defining the data points
x = [1, 2, 3, 4, 5]  # x-coordinates
y = [2, 5, 7, 8, 1]  # y-coordinates

# Creating the scatter plot
plt.scatter(x, y)

# Adding labels and title
plt.title("Scatter Plot of X vs Y")  # Title of the plot
plt.xlabel("X Values")  # Label for the x-axis
plt.ylabel("Y Values")  # Label for the y-axis

# # Optionally, you can add gridlines for better readability
# plt.grid(True)

# Displaying the plot
plt.show()  # This will show the plot window
```

```python
import matplotlib.pyplot as plt
x=[10,15,4,20,25,30]
y=[5,10,13,15,30,40]

plt.title("Swamy table")
plt.xlabel("Goods")
plt.ylabel("Exports")
plt.grid(False)
# plt.scatter(x,y)
plt.plot(x,y,color="green",marker="o",linestyle = '--',linewidth=2)
```

# Adding the multiple lines in it

```python

import matplotlib.pyplot as plt
x=[10,15,4,20,25,30]
y=[5,10,13,15,30,40]
y2=[15,30,43,85,10,12]
plt.title("Swamy table")
plt.xlabel("Goods")
plt.ylabel("Exports")
plt.grid()
# plt.scatter(x,y)
plt.plot(x,y,color="green",marker="o",linestyle = '--',linewidth=2)
plt.plot(x,y2,color="blue",marker="o",linestyle = '--',linewidth=2)
```

```python
# lines styles
plt.plot(x,np.sin(x),'--')
```

```python
import numpy as np  # Importing NumPy
import matplotlib.pyplot as plt  # Importing Matplotlib's pyplot module

# Generating 55 points between 0 and 10
x = np.linspace(0, 10, 55)

# Plotting the sine of each x value
plt.plot(x, np.sin(x), '-')  # '-' indicates a solid line

# Adding labels and title
plt.title("Sine Wave Plot")  # Title of the plot
plt.xlabel("X Values")  # X-axis label
plt.ylabel("sin(X)")  # Y-axis label

# Optionally, you can add a grid for better readability
plt.grid(True)

# Displaying the plot
plt.show()  # This will show the plot window
```
