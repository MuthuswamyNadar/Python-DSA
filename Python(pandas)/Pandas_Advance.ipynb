# Using Matplot lib


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# doing the basic plot
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# Customizing line style
plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth=2)
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='-',linewidth=2)
```

```python
plt.plot(x,y,label="sin(x)")
```

```python
plt.plot(x,np.cos(x),label="cos(x)",linestyle=":")
plt.grid(True)
plt.legend()
plt.show()
```
