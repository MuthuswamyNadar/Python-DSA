# Using Matplot lib


```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# doing the basic plot
x=np.linspace(0,10,50)
y=np.sin(x)
plt.plot(x,y)
plt.show()
```

```python
# Customizing line style
plt.plot(x,y,color='red',marker='o',linestyle='--',linewidth=2)
```

```python
plt.plot(x,y,color='red',marker='o',linestyle='-',linewidth=2)
```

```python
plt.plot(x,y,label="sin(x)")
```

```python
plt.plot(x,np.cos(x),label="cos(x)",linestyle=":")
plt.grid(True)
plt.legend()
plt.show()
```

# Subplots

```python
# plt.subplot(nrows, ncols, index)

x=[1,2,3,4,5]
y=[10,20,30,40,50]

plt.subplot(2,2,1)

plt.plot(x,y)
plt.title("sub1")

plt.subplot(2,2,2)

plt.plot(x,y)
plt.title("sub2")

```

```python
# Histogram
data=np.random.randn(1000)
plt.hist(data,bins=10)
plt.grid(True)
plt.show()
```

```python
datam=np.random.randn(100)
plt.hist(datam,bins=10)
plt.grid(True)
plt.show()
```

```python
#using the random 
a=np.random.rand(30)
b=np.random.rand(30)
plt.hist([a,b],bins=15,stacked=True)
plt.show()
```

```python
datam=np.random.rand(100)
dataz=np.random.rand(100)
plt.hist([datam,dataz],bins=30,stacked=True)
plt.grid(True)
plt.show()
```

```python
x=np.random.rand(10)
y=np.random.rand(10)
plt.scatter(x,y)
plt.show()
```

```python
# horizontal
categroy=['A','B','C']
values=[10,15,71]
plt.barh(categroy,values)
plt.show()
```

```python
#pie chart
size=[10, 30, 50]
labels=['A','B','C']
plt.pie(size,labels=labels,shadow=True,startangle=90)
plt.axis('equal')
plt.show()
```

```python
a=[10,20,30]
b=['swamy','Ammy','Zammy']
plt.pie(a,labels=b,shadow=True)
plt.grid(True)
plt.show()
```
